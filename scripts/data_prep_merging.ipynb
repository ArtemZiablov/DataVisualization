```python
import pandas as pd
from random import choices
```

```python
df = pd.read_csv('download.csv')
cities = pd.read_csv('filtered.csv')

df['dob.date'] = df['dob.date'].str.split('T').str[0]
df['gender'] = df['gender'].replace({'male': 'M', 'female': 'F'})
df['avatar'] = df['name.first'].str[0]
df = df.drop(columns=['dob.age'])

cities
```

```python
df
```

```python
c = choices(cities['feature_count'].tolist(), weights=cities['pop'], k=len(df))
```

```python
df['feature_count'] = c
cities_subset = cities[['feature_count', 'name',
                        'lat', 'lon', 'code', 'pop', 'timezone']]
df = df.merge(cities_subset, on='feature_count', how='left')
df.drop('feature_count', axis=1, inplace=True)
```

```python
df.to_csv('final.csv')
```
